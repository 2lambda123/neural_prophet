```python
import pandas as pd
from neuralprophet.neural_prophet import NeuralProphet
```

```python
df = pd.read_csv('../data/example_wp_log_peyton_manning.csv')
```

# Seasonality

##  Additive Seasonality

```python
m = NeuralProphet()
m.fit(df)
```

```python
forecast = m.predict(future_periods=m.n_forecasts)
fig_fit = m.plot(forecast)
```

```python
fig_trend = m.plot_components(forecast)
```

##  Multiplicative Seasonality
Clearly, additive seasonality is the better fit for this data.

```python
m = NeuralProphet(seasonality_mode='multiplicative')
m.fit(df)
```

```python
forecast = m.predict(future_periods=m.n_forecasts)
fig_fit = m.plot(forecast)
```

```python
fig_trend = m.plot_components(forecast)
```

##  Underfit Additive Seasonality

```python
m = NeuralProphet(
    yearly_seasonality=2,
    weekly_seasonality=1,
    daily_seasonality=False,
)
m.fit(df)
```

```python
forecast = m.predict(future_periods=m.n_forecasts)
fig_fit = m.plot(forecast)
```

```python
fig_trend = m.plot_components(forecast)
```

##  Overfit Additive Seasonality

```python
m = NeuralProphet(
    yearly_seasonality=16,
    weekly_seasonality=8,
    daily_seasonality=False,
)
m.fit(df)
```

```python
forecast = m.predict(future_periods=m.n_forecasts)
fig_fit = m.plot(forecast)
```

```python
fig_trend = m.plot_components(forecast)
```
