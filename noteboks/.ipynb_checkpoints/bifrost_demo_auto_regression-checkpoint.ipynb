```python
import pandas as pd
from neuralprophet.neural_prophet import NeuralProphet as Bifrost
```

```python
df = pd.read_csv('../data/example_wp_log_peyton_manning.csv')
```

# Auto-Regression

## AR(1): 1 step ahead

```python
m = Bifrost(
    n_forecasts=1,
    n_lags=1,
    yearly_seasonality=False,
    weekly_seasonality=False,
    daily_seasonality=False,
)
m.fit(df)
```

```python
forecast = m.predict(future_periods=m.n_forecasts, freq='D')
fig_fit = m.plot(forecast)
```

```python
fig_trend = m.plot_components(forecast)
# fig_prediction = m.plot(forecast, crop_last_n=m.n_lags+m.n_forecasts)
# single_forecast = m.get_last_forecasts(10)
# fig_prediction = m.plot(single_forecast)
```

### AR(60): 1 step ahead

```python
m = Bifrost(
    n_forecasts=1,
    n_lags=60,
    yearly_seasonality=False,
    weekly_seasonality=False,
    daily_seasonality=False,
)
m.fit(df)
```

```python
forecast = m.predict(future_periods=m.n_forecasts, freq='D')
fig_fit = m.plot(forecast)
# fig_trend = m.plot_components(forecast)
fig_prediction = m.plot(forecast, crop_last_n=m.n_lags+m.n_forecasts)
# single_forecast = m.get_last_forecasts(10)
# fig_prediction = m.plot(single_forecast)
```

### AR(60): 1-30 step ahead

```python
m = Bifrost(
    n_forecasts=30,
    n_lags=60,
    yearly_seasonality=False,
    weekly_seasonality=False,
    daily_seasonality=False,
)
m.fit(df)
```

```python
forecast = m.predict(future_periods=m.n_forecasts, freq='D')
fig_fit = m.plot(forecast)
# fig_trend = m.plot_components(forecast)
```

```python
fig_prediction1 = m.plot(forecast, crop_last_n=m.n_lags+m.n_forecasts, highlight_forecast=1)
fig_prediction2 = m.plot(forecast, crop_last_n=m.n_lags+m.n_forecasts, highlight_forecast=m.n_forecasts)
```

```python
single_forecast = m.get_last_forecasts(5)
fig_prediction = m.plot(single_forecast)
```

### AR(60): 1-30 step ahead, Sparse weights

```python
m = Bifrost(
    n_forecasts=30,
    n_lags=60,
    ar_sparsity=0.01,
    yearly_seasonality=False,
    weekly_seasonality=False,
    daily_seasonality=False,
)
m.fit(df)
```

```python
forecast = m.predict(future_periods=m.n_forecasts, freq='D')
fig_fit = m.plot(forecast)
# fig_trend = m.plot_components(forecast)
```

```python
fig_prediction1 = m.plot(forecast, crop_last_n=m.n_lags+m.n_forecasts, highlight_forecast=1)
fig_prediction2 = m.plot(forecast, crop_last_n=m.n_lags+m.n_forecasts, highlight_forecast=m.n_forecasts)
```

```python
single_forecast = m.get_last_forecasts(5)
fig_prediction = m.plot(single_forecast)
```
