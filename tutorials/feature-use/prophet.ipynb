```python
import pandas as pd

data_location = "https://raw.githubusercontent.com/ourownstory/neuralprophet-data/main/datasets/"
df = pd.read_csv(data_location + "wp_log_peyton_manning.csv")
```

```python
from prophet import Prophet
p = Prophet()

from neuralprophet.wrapper import Prophet
np = Prophet()
```

```python
p.fit(df)
np.fit(df)
```

```python
future_p = p.make_future_dataframe(periods=365)
future_p
```

```python
future_np = np.make_future_dataframe(df=df, periods=365)
future_np
```

```python
forecast_p = p.predict(future_p)
forecast_p
```

```python
forecast_np = np.predict(future_np)
forecast_np
```

```python
fig1_p = p.plot(forecast_p)
```

```python
fig1_np = np.plot(forecast_np)
```

```python
fig2_p = p.plot_components(forecast_p)
```

```python
fig2_np = np.plot_components(forecast_np)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

plot_plotly(p, forecast_p)
```

```python
np.plot(forecast_np, plotting_backend="plotly")
```

```python
plot_components_plotly(p, forecast_p)
```

```python
np.plot_components(forecast_np, plotting_backend="plotly")
```
